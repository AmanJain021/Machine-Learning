```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.datasets import mnist
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, confusion_matrix
import seaborn as sns
# Step 1: Load the dataset
(X_train, y_train), (X_test, y_test) = mnist.load_data()
# Step 2: Flatten 28x28 images into 1D feature vectors of size 784
X_train = X_train.reshape(X_train.shape[0], -1)
X_test = X_test.reshape(X_test.shape[0], -1)
# Step 3: Normalize pixel values (scale from 0-255 to 0-1)
X_train = X_train / 255.0
X_test = X_test / 255.0
# Step 4: Print shape to confirm transformation
print(f"Training Data Shape: {X_train.shape}, Test Data Shape: {X_test.shape}")
# Step 5: Train an SVM model with RBF kernel
svm_model = SVC(kernel='rbf', gamma='scale', C=10)
# C controls the regularization
print("Training SVM model...")
svm_model.fit(X_train[:10000], y_train[:10000])
# Train on a subset (10,000 samples) to reduce computation time
print("Model training completed!")
# Step 6: Make predictions on test data
y_pred = svm_model.predict(X_test)
# Step 7: Calculate Accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Model Accuracy: {accuracy:.4f}")
# Step 8: Display Confusion Matrix
plt.figure(figsize=(10, 6))
sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, fmt="d", cmap="Blues")
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.title("Confusion Matrix")
plt.show()
# Step 9: Display sample images with their predictions
fig, axes = plt.subplots(3, 5, figsize=(10, 6))
for i, ax in enumerate(axes.flat):
    ax.imshow(X_test[i].reshape(28, 28), cmap="gray")
    ax.set_title(f"Pred: {y_pred[i]}\nTrue: {y_test[i]}")
    ax.axis("off")
plt.tight_layout()
plt.show()
```
